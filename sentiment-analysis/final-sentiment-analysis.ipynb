# Imports/Setup

```python
import json

import nltk
from textblob import TextBlob
from textblob.classifiers import NaiveBayesClassifier
from cleantext import clean
```

# Train Naive Bayes Classifier

```python
#training data
with open('trainfile.csv', 'r') as fp:
    cl = NaiveBayesClassifier(fp, format="csv")
```

```python
with open('testfile.csv', 'r') as test:
    print(cl.accuracy(test, format="csv"))
```

# Classify entire dataset

```python
# Opening JSON file
f = open('comments.json')
# returns JSON object as 
# a dictionary
comment = json.load(f)

#dictionary to return scores
scoresDict = {
    "Lewis": [],
    "Hamilton": [],
    "Max": [],
    "Verstappen": [],
    "Valtteri": [],
    "Bottas": [], 
    "Charles": [],
    "Leclerc": [], 
    "Daniel": [], 
    "Ricciardo": [],
    "Lando": [],
    "Norris": [], 
    "Carlos": [],
    "Sainz": [], 
    "Sergio": [],
    "Perez": [], 
    "Fernando": [],
    "Alonso": [], 
    "Esteban": [],
    "Ocon": [], 
    "Pierre": [],
    "Gasly": [], 
    "Yuki": [],
    "Tsunoda": [], 
    "Oscar": [],
    "Piastri": [],
    "Zhou": [], 
    "Guanyu": [],
    "Logan": [], 
    "Sargeant": [],
    "Alex": [], 
    "Albon": [],
    "George": [],
    "Russell": [], 
    "Kevin": [], 
    "Magnussen": [],
    "Nico": [], 
    "Hulkenberg": [],
    "Lance": [],
    "Stroll": [],
    "Race Discussion": []
}
```

```python
sentences = []

for driver in comment:
    for entry in comment[driver]:
        sentences.append(clean(entry[0], no_emoji=True, no_line_breaks=True, fix_unicode=True, no_urls=True, no_numbers=True))
        scoresDict[driver].append([entry[1], entry[2], entry[3]])
```

```python
sentences
```

```python
scoreNums = []

for s in sentences:
    if cl.classify(s) == "pos":
        scoreNums.append(1)
    else:
        scoreNums.append(-1)
```

```python
#combine scores with other metrics
index = 0
for driver in scoresDict:
    for entry in scoresDict[driver]:
        entry.append(scoreNums[index])
        index+=1
```

```python
# Save comments to a JSON file
with open("scores_final.json", "w") as outfile:
    json.dump(scoresDict, outfile, indent=4)
```
