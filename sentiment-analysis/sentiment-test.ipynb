```python
sentences = [
    "if they let go of yuki he'll bite",
    "My prediction. He was mean and vulgar about a work issue through text to a female employee. Someone that doesn't like Christian took this opportunity and made a circus out of it.",
    "Who did they hire to take the shot of Liam Lawson. Got dang they did him dirty, guy looks 12 in that shot.",
    "I love lewis",
    "bro looks like the average middle school bully in that outfit",
    "The top looks great, very 90s. I love it.",
    "Sounds like he's fairly confident they have a much better car than last year.",
    "bad",
    "i'm sorry but vcarb looks shit"
]
```

```python
for sentence in sentences:
    sid = SentimentIntensityAnalyzer()
    print(sentence)
    ss = sid.polarity_scores(sentence)
    for k in sorted(ss):
        print('{0}: {1}, '.format(k, ss[k]), end='')
    print("\n")
```
